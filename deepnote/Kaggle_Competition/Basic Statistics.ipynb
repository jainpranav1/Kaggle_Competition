# Basic Statistics

```
# import the libraries
import dask.dataframe as dd
import matplotlib.pyplot as plt
```

```
# import and see train_data.csv
df_train_csv = dd.read_csv('Data/train_data.csv')
df_train_csv.head()
```

```
# split into 2A3 MaP and DMS MaP datasets
df_2A3 = df_train_csv[df_train_csv["experiment_type"]=="2A3_MaP"]
df_DMS = df_train_csv[df_train_csv["experiment_type"]=="DMS_MaP"]
```

```
df_2A3.head()
```

```
df_DMS.head()
```

```
# get number of sequences
df_2A3.shape[0].compute()
```

```
# get number of unique sequences
df_2A3["sequence"].drop_duplicates().shape[0].compute()
```

```
# get sequence token lengths and counts
df_2A3["sequence"].str.len().value_counts().compute()
```

```
# get dataset name and counts
df_2A3["dataset_name"].value_counts().compute()
```

```
# import and see test_sequences.csv
df_test_csv = dd.read_csv('Data/test_sequences.csv')
df_test_csv.head()
```

```
# get number of sequences
df_test_csv.shape[0].compute()
```

```
# get number of unique sequences
df_test_csv["sequence"].drop_duplicates().shape[0].compute()
```

```
# get sequence lengths and counts
df_test_csv["sequence"].str.len().value_counts().compute()
```
